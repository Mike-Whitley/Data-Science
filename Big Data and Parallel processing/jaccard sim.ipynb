```
def jaccard_similarity(list1, list2):
    """calculates the jaccard similarity between two lists can be done on two words
    just change each item in list to single letter"""
    s1 = set(list1)
    s2 = set(list2)
    return len(s1.intersection(s2)) / len(s1.union(s2))

name_a = ['J', 'O', 'H', 'N']
#name_b = ['T', 'E', 'R','R','Y']
name_b = ['J', 'O', 'S', 'H']
print(jaccard_similarity(name_a, name_b))
```
